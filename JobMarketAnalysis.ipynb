```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

# Load the dataset
file_path = "Job_listing.csv"  # Change if needed
df = pd.read_csv(file_path)

# Display the first few rows
print("📌 First 5 job listings:")
display(df.head())

# Count the most common job titles
top_jobs = df["Job"].value_counts().head(10)

# Display the most popular job titles
print("\n📊 Top 10 most common job titles:")
print(top_jobs)

# Plot the most common job titles
plt.figure(figsize=(10, 5))
top_jobs.plot(kind="bar", color="skyblue")
plt.title("Top 10 Most Common Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Extract job titles for skill analysis
job_titles = " ".join(df["Job"].dropna())

# Define common tech-related keywords
keywords = ["Python", "SQL", "AWS", "Cloud", "Security", "Machine Learning",
            "BI", "DevOps", "Docker", "Kubernetes", "Java", "JavaScript", 
            "ETL", "Analytics", "Big Data", "TensorFlow", "NLP"]

# Count occurrences of each keyword in job titles
keyword_counts = Counter({word: len(re.findall(word, job_titles, re.IGNORECASE)) for word in keywords})

# Sort by most mentioned skills
top_skills = dict(sorted(keyword_counts.items(), key=lambda item: item[1], reverse=True))

# Display the top skills
print("\n🔍 Top mentioned skills in job titles:")
print(top_skills)

# Plot the skill distribution
plt.figure(figsize=(10, 5))
plt.bar(top_skills.keys(), top_skills.values(), color="orange")
plt.title("Most Mentioned Skills in Job Titles")
plt.xlabel("Skills")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
